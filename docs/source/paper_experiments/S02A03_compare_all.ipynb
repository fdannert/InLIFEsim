## Compare yield numbers

```python
yields = pd.DataFrame(columns=['number', 'noise', 'stype', 'habitable'], index=np.arange(24))
stype = ['F', 'G', 'K', 'M']

for i in range(1, 5):
    yields.number.iloc[i+7] = bus_0.data.catalog.detected[bus_0.data.catalog.stype==i].sum()/500
    yields.noise.iloc[i+7] = 'none'
    yields.stype.iloc[i+7] = stype[i-1]
    yields.habitable.iloc[i+7] = False
    
    yields.number.iloc[i-1] = bus_1.data.catalog.detected[bus_1.data.catalog.stype==i].sum()/500
    yields.noise.iloc[i-1] = 'lay'
    yields.stype.iloc[i-1] = stype[i-1]
    yields.habitable.iloc[i-1] = False
    
    yields.number.iloc[i+3] = bus_2.data.catalog.detected[bus_2.data.catalog.stype==i].sum()/500
    yields.noise.iloc[i+3] = 'earth twin'
    yields.stype.iloc[i+3] = stype[i-1]
    yields.habitable.iloc[i+3] = False
    
    yields.number.iloc[i+19] = bus_0.data.catalog.detected[
        np.logical_and(bus_0.data.catalog.stype==i, bus_0.data.catalog.habitable)].sum()/500
    yields.noise.iloc[i+19] = 'none'
    yields.stype.iloc[i+19] = stype[i-1]
    yields.habitable.iloc[i+19] = True
    
    yields.number.iloc[i+11] = bus_1.data.catalog.detected[
        np.logical_and(bus_1.data.catalog.stype==i, bus_1.data.catalog.habitable)].sum()/500
    yields.noise.iloc[i+11] = 'lay'
    yields.stype.iloc[i+11] = stype[i-1]
    yields.habitable.iloc[i+11] = True
    
    yields.number.iloc[i+15] = bus_2.data.catalog.detected[
        np.logical_and(bus_2.data.catalog.stype==i, bus_2.data.catalog.habitable)].sum()/500
    yields.noise.iloc[i+15] = 'earth twin'
    yields.stype.iloc[i+15] = stype[i-1]
    yields.habitable.iloc[i+15] = True
```

```python
def barplot_dev(bins, param, ax, ret=False):
    actbins = np.linspace(bus_0.data.catalog[bus_0.data.catalog.detected][param].min(), 
                          bus_0.data.catalog[bus_0.data.catalog.detected][param].max(), bins)
    actbins_mid = actbins[:-1] + (actbins[1:]-actbins[:-1])/2

    height = ((np.histogram(bus_0.data.catalog[bus_0.data.catalog.detected][param], bins=actbins)[0] 
             - np.histogram(bus_1.data.catalog[bus_1.data.catalog.detected][param], bins=actbins)[0])
            / (np.histogram(bus_0.data.catalog[bus_0.data.catalog.detected][param], bins=actbins)[0] 
               + np.histogram(bus_1.data.catalog[bus_1.data.catalog.detected][param], bins=actbins)[0]))
    
    height[height == 1.] = np.max(height[height != 1.])
    height[height == -1.] = np.min(height[height != -1.])
    
    ax.bar(actbins_mid, 
           height,
            alpha=0.8, 
            width=actbins[1:]-actbins[:-1])
    ax.set_xlabel(param)
    ax.set_ylabel('Excess in base noise')
    
    if ret:
        return height
```
